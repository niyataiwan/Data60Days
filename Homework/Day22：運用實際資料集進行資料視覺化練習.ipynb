   在帶狀圖中，散點通常將重疊，將無法觀察到數據的完整分布。
   一個簡單的解決方案是使用一些隨機的「抖動」調整位置（僅沿著分類軸），抖動是平時可視化中的常用的觀察「密度」的方法。

```python
sns.stripplot(x = "species", y = "petal_length", data = iris, jitter = True)
#jitter = True, False 代表抖動量高、抖動量低
```

```python
sns.stripplot(x = "species", y = "petal_length", data = iris, jitter = 0.05)
```

3. 聯合分佈圖sns.pairplot()

pairplot 圖能直觀的反映出兩兩特徵間的關係，幫助我們對數據集建立初步印象，更好的完成分類和聚類任務。

```python
g = sns.pairplot(iris, hue = "species", diag_kind = "kde", kind = "scatter", palette = "husl")
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
g.map_diag(sns.kdeplot, lw = 3)

# map_diag（func，** kwargs）：在每個對角線子圖上繪製具有單變數函數的圖。
# map_lower（func，** kwargs）：在下對角線子圖上繪製具有雙變數函數的圖。
# map_upper（func，** kwargs）：在上對角線子圖上繪製具有雙變數函數的圖
# map_offdiag（func，** kwargs）：在非對角線子圖上繪製具有雙變數函數的圖。
```

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
titanic = sns.load_dataset("titanic")
```

```python
titanic.head()
```

```python
titanic.info()
```

1. 將每個class中男性和女性的平均存活率用長形圖可視化

```python
sns.barplot(x = "sex", y = "survived", hue = "class", data = titanic)
# class: first, second, third
# sex: Male, Female
#survived: number
```

2.瞭解性別在各艙等存活率的分布

```python
g = sns.FacetGrid(titanic, col = "pclass") #生成3個座標格，表示plcass有三類
g.map(sns.barplot, "sex", "survived", order = ["male", "female"])

#class 1,2,3的female存活率高於male，且class 1的female存活率最高
```

3.繪製堆疊條形圖，x軸代表依據艙等分成男性及女性，y軸代表人數，其中藍色代表死亡人數，橘色代表存活人數

```python
#先檢視各艙位存活人數，可以使用groupby分類
#其中survived = 0, 1表示死亡和存活，將各艙的survived加總，即為存活人數

#加上性別
survived = titanic.groupby(["pclass", "sex"]).survived.sum()
print(survived)
#survived.plot(kind = 'bar')
```

```python
#用pd.crosstab函數繪製交叉表，交叉表可以直觀的依據艙位等級和性別查看存活人數和死亡人數
#繪製堆疊條形圖，x軸依據艙等分男性和女性，y為人數，其中藍色是死亡人數，橘色是存活人數
survived_counts = pd.crosstab([titanic.pclass, titanic.sex], titanic.survived)
survived_counts.index
```

```python
survived_counts.plot.bar(stacked = True)
```

4. 瞭解性別在各艙等的存活率的分布

```python
g = sns.FacetGrid(titanic, col = "sex")
g.map(sns.barplot, "pclass", "survived", order = [1,2,3])
```
