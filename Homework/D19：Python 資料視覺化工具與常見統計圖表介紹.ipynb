2-1. 散點圖 Scatter Plots: 給定顏色和圖形形狀
     數值介於0~10 array, 間隔0.4
     畫出y = x, y = x平方, y= x 三次方的圖表
     r代表紅色 ; 兩個-代表虛線
     b代表藍色 ; s代表方塊
     g代表綠色 ; ^代表三角形

```python
t = np.arange(0, 10, 0.4)
print("t: "+str(t))

plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
#t, t, 'r--'  : y 紅色虛線
#t, t**2, 'bs': y 平方 藍色方塊
#t, t**3, 'g^': y 三次方 綠色三角形
```

2-1-1. 增加圖示說明，搭配legend()

```python
t = np.arange(0, 10, 0.4)
print("t: "+ str(t))

plt.plot(t, t   , 'r--', label = "y" )
plt.plot(t, t**2, 'bs' , label = "y2")
plt.plot(t, t**3, 'g^' , label = "y3")
plt.legend()
```

4. 進階-善用series and DataFrame的plot繪製圖表

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

4-1. 單一折線圖

```python
ts = pd.Series(np.random.randn(200), index = pd.date_range('1/1/2023', periods = 200))
ts = ts.cumsum()
ts.plot()

plt.title('Random Data')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend(['Cumulative Sum'])

plt.show()
```

補充:
1. np.random.randn()
   隨機得到一個數字，可能是負的或正的，可能大於一或小於一，總之就是隨機
2. numpy.cumsum(a, axis=None, dtype=None, out=None)
   axis=0，按照行累加。
   axis=1，按照列累加。
   axis不給定具體值，就把numpy數組當成一個一維數組

4-2. 多個折線圖

```python
df = pd.DataFrame(np.random.randn(200, 3), index = pd.date_range('1/1/2023', periods = 200), 
                  columns = ["A", "B", "C"])
df = df.cumsum()
print(df)
df.plot();
```

1. 畫出 cos 圖檔，並儲存

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 180)
y = np.cos(x * np.pi / 180.0)

plt.plot(x, y)
plt.show()  #畫出 cos 圖檔

#畫出完整的 cos 圖形
x = np.arange(0, 3 * np.pi, 0.1)
y_cos = np.cos(x)
plt.plot(x, y_cos)
#plt.show()

plt.savefig("cos.png",dpi=300,format="png") #儲存
```

```python
n = 250 #sample數量
X = np.random.normal(0, 1, n)
Y = np.random.normal(0, 1, n)

T = np.arctan2(X, Y)
plt.scatter(X, Y, s=75, c=T, alpha =.6) #s: 點的大小; alpha: 透明度; c: 一維numpy array

plt.title("Scatter Plots")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)

plt.show()
T
```
